```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
markers = ['o', 'v', 's', 'X', 'd', '*', 'p', '>']
```

```python
csv_files = ['Tests/noisy_function/csv_files/runge_mse_0.01.csv',
             'Tests/noisy_function/csv_files/abs_mse_0.01.csv',
             'Tests/noisy_function/csv_files/tan_mse_0.01.csv']
```

```python
plt.figure(figsize=(12, 8))

for i, file in enumerate(csv_files):
    df = pd.read_csv(file, index_col=0)
    
    plt.subplot(2, 2, i + 1)
    
    j = 0
    for col in df.columns:
        if markers[j] == '*':
            plt.plot(np.sqrt(df[col]), marker=markers[j], markersize=8, linewidth=1)
        else:
            plt.plot(np.sqrt(df[col]), marker=markers[j], markersize=6, linewidth=1)
        j += 1
        
    plt.yscale('log')
    
    if i == 0:
        plt.ylim(5e-3, 1)
        plt.ylabel('$\|f - r\|_{2}$', fontsize=15)
        plt.xticks(np.arange(5, 41, 5), ['' for _ in np.arange(5, 41, 5)])
        plt.text(33, 0.03, r"(A) $\frac{1}{1+(10(x - 0.5))^2}$", 
                 fontsize=14, horizontalalignment='center', verticalalignment='center')
    elif i == 1:
        plt.ylim(5e-3, 1)
        plt.xlabel('$n$', fontsize=15)
        plt.text(8, 7e-3, r"(B) $|10(x - 0.5)|$", 
                 fontsize=14, horizontalalignment='center', verticalalignment='center')
    elif i == 2:
        plt.ylim(5e-3, 1)
        plt.ylabel('$\|f - r\|_{2}$', fontsize=15)
        plt.xlabel('$n$', fontsize=15)
        plt.text(10, 7e-3, r"(C) $\arctan(50(x - 0.5))$", 
                 fontsize=14, horizontalalignment='center', verticalalignment='center')
            
    i += 1
    
fig = plt.subplot(2, 2, 4)
for j, col in enumerate(df.columns):
    if col in ['Bernstein', 'SK', 'Linearized Bernstein']:
        label = f"(n, n) {col}" if col != 'SK' else f"(n, n) Sanathan-Koerner"
    elif col == 'Polynomial':
        label = f"2n Degree Polynomial"
    elif col == 'Poleless Barycentric':
        label = f"(n+1, n+1) (d=3) {col}"
    elif col == 'Natural Spline':
        label = f"(df=2n+1) {col}"
    else:
        label = f"(n+1, n+1) {col}"
    
    if markers[j] == '*':
        plt.plot([], [], marker=markers[j], markersize=8, label=label)
    else:
        plt.plot([], [], marker=markers[j], markersize=6, label=label)
    
    plt.legend(loc='center')
    
    plt.axis('off')
    
plt.tight_layout()
plt.savefig('small_noise_function_convergence.png', dpi=150)
```
