```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
markers = ['o', 'v', 's', 'X', 'd', '8', 'p', '>']
```

```python
fig, ax = plt.subplots(1, 1, figsize=(7, 4))

df = pd.read_csv('Tests/noisy_function/csv_files/abs_mse_0.01.csv', index_col=0)

i = 0
for col in df.columns:
    if col in ['Linearized Bernstein', 'Natural Spline']:
        continue
    
    if col in ['Bernstein', 'SK']:
        label = f"(n, n) {col}" if col != 'SK' else f"(n, n) Sanathan-Koerner"
    elif col == 'Polynomial':
        label = f"2n Degree Polynomial"
    else:
        label = f"(n+1, n+1) {col}"
    
    ax.plot(df[col], label=label, marker=markers[i])
    i += 1
    
ax.set_yscale('log')
ax.set_ylim(5e-5, 5)
ax.legend()

ax.set_ylabel('MSE', fontsize=15)
ax.set_xlabel('n', fontsize=15)

plt.tight_layout()
plt.savefig(f"noisy_abs_convergence.png", dpi=150)
```
